```python
gcc= load_catalog('proto-dc2_v4.6.1')
```

```python
gcc.cosmology
```

```python
data = pd.DataFrame(gcc.get_quantities(['galaxy_id', 'stellar_mass', 'redshift','redshift_true','totalStarFormationRate','SDSS_filters/magnitude:SDSS_g:observed','dec_true','ra_true'],filters = ['redshift <.2' ]))
```

```python
data=data.sort_values(by=['stellar_mass'])

```

```python
data2=data.set_index("galaxy_id", drop = False)
data2
```

```python
data2['SNR']=(.26e-10*(data2['stellar_mass']**(.72)))+(1.23e-03*(data2['totalStarFormationRate']**(.95)))
data2=data2.sort_values(by=['SNR'])
data2
```

```python
data2['SNR_cumsum']=data2.SNR.cumsum()
```

```python
data2['Norm']=data2['SNR_cumsum']/data2.SNR_cumsum.max()
data2
```

```python
def Cal_SNR(row):
    return (.26e-10)*row['stellar_mass']**(.72)+1.23e-3*row['totalStarFormationRate']**(.95)
```

```python
data['SNR']=data.apply(Cal_SNR, axis=1)
```

```python
data=data.sort_values(by=['SNR'])
data
```

```python
def Cal_Cuum(column):
    return np.cumsum(column['SNR'])
```

```python
data1=data.apply(Cal_Cuum,axis=1)
data1
```

```python
def Div_by(row):
    return row['SNR']/6.636820e+06
```

```python
data2['norm_SNR']=pd.DataFrame(data1.apply(Div_by,axis=1))
#data2= data2.drop(columns='norm')
#data2=data2.drop(columns='(g_id, norm_SNR)')
data2
```
